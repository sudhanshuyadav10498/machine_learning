### SVD: X = U · Σ · VT

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.image import imread


X = imread('monkey.png')

def show_image(img):
    plt.imshow(img)
    plt.axis('off')
    plt.show()
```

```python
print('RBG image')
show_image(X)

print('Gray image')
Xgray = np.mean(X, -1)
show_image(Xgray)
```

```python
U, s, VT = np.linalg.svd(Xgray,full_matrices=False)
S=np.diag(s)

# Just taking first 150 eigen samples from all available samples
r=100 
Xapprox = U @ S @ VT 

print('COmpressed Image - Reconstructed with 150 eigens')
show_image(Xapprox)
```

```python
# Image compression with first r eigens

for r in (5, 50, 75, 100, 200):
    Xapprox = U[:,:r] @ S[:r,:r] @ VT[:r,:]
    print('First ',r,' eigens featues')    
    show_image(Xapprox)

```

For r > 75 seems adequately compressed from passed values
